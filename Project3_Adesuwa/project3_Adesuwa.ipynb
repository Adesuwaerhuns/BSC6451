Problem 1
Write a script (or Jupyter Notebook code block) that opens the file, uses a for loop to read through the file line by line and reports the highest water level and the date and time that was observed.

```python
waterLevel_list = []
dateTime_list = []
with open('proj3.csv','r')as fh:
    for line in fh.readlines():
        dateTime_list.append((line.split(','))[0])
        waterLevel_list.append((line.split(','))[1])
# Removing the header in the list
waterLevel = waterLevel_list[1:]
dateTime = dateTime_list[1:]
#This code cleans up the data by removing the nan value in index 423 
waterLevel = waterLevel_list[1:]
dateTime = dateTime_list[1:]
del waterLevel[423] 
del dateTime[423]
#make the waterlist a float type to enable us perform mathematical operations
waterLevel=[float(i) for i in waterLevel]
print( "Highest water level is -" ,max(waterLevel))
print( "Date and Time stamp is - ", dateTime[(waterLevel.index(max(waterLevel)))])
```
